```python
import torch
from collections import OrderedDict

ckpt = torch.load('{}/{}'.format(pre_trained_model_path, pre_trained_model_name), map_location='cpu')
state_dict = ckpt.get('model_state_dict', ckpt)  # 兼容直接保存 state_dict 的情况

def rewrite_keys_1(sd):
    new_sd = OrderedDict()
    for k, v in sd.items():
        k2 = k
        k2 = k2.replace('encoder.ComENet_1.', 'encoder.GM_1.')
        k2 = k2.replace('encoder.ComENet_2.', 'encoder.GM_2.')
        for i in range(16):
            k2 = k2.replace(f'encoder.Graph_Norm_{i}.', f'encoder.Global_Norm_{i}.')

        k2 = k2.replace('decoder.Folding_Unit_1.', 'decoder.FO_1.')
        k2 = k2.replace('decoder.Folding_Unit_2.', 'decoder.FO_2.')
        k2 = k2.replace('decoder.Folding_Unit_3.', 'decoder.FO_3.')

        new_sd[k2] = v
    return new_sd


def rewrite_keys_2(sd):
    new_sd = OrderedDict()
    for k, v in sd.items():
        k2 = k
        k2 = k2.replace('decoder.FO_1.PredXYZ_1.', 'decoder.FO_1.PG_1.')
        k2 = k2.replace('decoder.FO_1.PredXYZ_2.', 'decoder.FO_1.PG_2.')
        k2 = k2.replace('decoder.FO_2.PredXYZ_1.', 'decoder.FO_2.PG_1.')
        k2 = k2.replace('decoder.FO_2.PredXYZ_2.', 'decoder.FO_2.PG_2.')
        k2 = k2.replace('decoder.FO_3.PredXYZ_1.', 'decoder.FO_3.PG_1.')
        k2 = k2.replace('decoder.FO_3.PredXYZ_2.', 'decoder.FO_3.PG_2.')

        new_sd[k2] = v
    return new_sd


state_dict = rewrite_keys_1(state_dict)
rewritten = rewrite_keys_2(state_dict)


pre_trained_model_state['model_state_dict'] = rewritten
model.load_state_dict(pre_trained_model_state['model_state_dict'], strict=True)

# # （可选）只保留与当前模型“键名一致且形状相同”的参数，避免形状不匹配
# model_sd = model.state_dict()
# filtered = {k:v for k,v in rewritten.items() if (k in model_sd and v.shape == model_sd[k].shape)}

# missing, unexpected = model.load_state_dict(filtered, strict=False)
# print('[Load ckpt] missing:', missing)
# print('[Load ckpt] unexpected:', unexpected)
```
